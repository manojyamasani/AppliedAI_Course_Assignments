<font color='red'><b>Step - 2 </b></font>

<font color='blue'><b>Building High Variance Models on each of the sample and finding train MSE value</b></font>

*  Build a regression trees on each of 30 samples.
*  Computed the predicted values of each data point(506 data points) in your corpus.
*  Predicted house price of $i^{th}$ data point $y^{i}_{pred} =  \frac{1}{30}\sum_{k=1}^{30}(\text{predicted value of } x^{i} \text{ with } k^{th} \text{ model})$
*  Now calculate the $MSE =  \frac{1}{506}\sum_{i=1}^{506}(y^{i} - y^{i}_{pred})^{2}$

<font color='red'> <b>Step - 3 </b></font>

*  <font color='blue'><b>Calculating the OOB score </b></font>

*  Predicted house price of $i^{th}$ data point $y^{i}_{pred} =  \frac{1}{k}\sum_{\text{k= model which was buit on samples not included } x^{i}}(\text{predicted value of } x^{i} \text{ with } k^{th} \text{ model})$.
*  Now calculate the $OOB Score =  \frac{1}{506}\sum_{i=1}^{506}(y^{i} - y^{i}_{pred})^{2}$.

*  <font color='blue'><b> Write code for generating samples</b></font>

```
def generating_samples(input_data, target_data):

    '''In this function, we will write code for generating 30 samples '''
    # you can use random.choice to generate random indices without replacement
    # Please have a look at this link https://docs.scipy.org/doc/numpy-1.16.0/reference/generated/numpy.random.choice.html for more details
    # Please follow above pseudo code for generating samples 
    
    #print(np.floor(0.6*len(input_data)))
    select_rows = np.random.choice(len(input_data),size = int(np.floor(0.6*len(input_data))),replace = False)
    #print(select_rows,select_rows.shape)
    replace_rows = np.random.choice((select_rows),size = int(np.ceil(0.4*len(input_data))),replace = False)
    #print(replace_rows,replace_rows.shape)

    no_col = 8
    select_cols = np.random.choice(len(input_data[0,:]),size =no_col, replace = False)
    #print(select_cols)
    sam = np.meshgrid(select_cols,select_rows)
    #print((sam[0].shape))
    #print(sam[1].shape)    

    sam_data = input_data[sam[1],sam[0]]
    #print(sam_data.shape)

    target_sam_data = target_data[select_rows]
    #print(target_sam_data.shape)

    sam = np.meshgrid(select_cols,replace_rows)
    #print((sam[0].shape))
    #print(sam[1].shape)
    replicate_data = input_data[sam[1],sam[0]]
    #print(replicate_data.shape)

    target_rep_data = target_data[replace_rows]
    #print(target_rep_data.shape)

    final_sam_data = np.concatenate((sam_data,replicate_data),axis = 0)
    final_target_data  = np.concatenate((target_sam_data,target_rep_data))

    #print(final_sam_data.shape,final_target_data.shape,select_rows.shape,select_cols.shape)

    return list(final_sam_data) , list(final_target_data),list(select_rows),list(select_cols)
    #note please return as lists
```

<font color='cyan'> <b> Grader function - 1 </b> </fongt>

```
def grader_samples(a,b,c,d):
    length = (len(a)==506  and len(b)==506)
    sampled = (len(a)-len(set([str(i) for i in a]))==203)
    rows_length = (len(c)==303)
    column_length= (len(d)>=3)
    assert(length and sampled and rows_length and column_length)
    return True
a,b,c,d = generating_samples(x, y)
grader_samples(a,b,c,d)
```

*  <font color='blue'> <b>Create 30 samples </b>

![alt text](https://i.imgur.com/p8eZaWL.jpg)

```
# Use generating_samples function to create 30 samples 
# store these created samples in a list
list_input_data =[]
list_output_data =[]
list_selected_rows= []
list_selected_columns=[]
no_base_models = 30
count = 0
i=0
while (i != (no_base_models+count)):
  a,b,c,d = generating_samples(x, y)
  d = sorted(d)
  i = i+1
  if ((d in list_selected_columns)):
    #print(count)
    count = count+1
    continue
  list_selected_columns.append(d)
  list_input_data.append(a)
  list_output_data.append(b)
  list_selected_rows.append(c)

  
```

```
def grader_30(a):
    assert(len(a)==30 and len(a[0])==506)
    return True
grader_30(list_input_data)
```

<font color='red'><b>Step - 2 </b></font>

<font color='orange'><b>Flowchart for building tree</b></font>

*  <font color='blue'><b> Write code for building regression trees</b></font>

```
from sklearn.tree import DecisionTreeRegressor

list_all_models = []
for i in range(no_base_models):
  input_data = list_input_data[i]
  target_data = list_output_data[i]
  model =  DecisionTreeRegressor()
  model.fit(input_data, target_data)
  list_all_models.append(model)

#print(list_all_models)
```

<font color='orange'><b>Flowchart for calculating MSE </b></font>

![alt text](https://i.imgur.com/sPEE618.png)

After getting predicted_y for each data point, we can use sklearns mean_squared_error to calculate the MSE between predicted_y and actual_y.

*  <font color='blue'><b> Write code for calculating MSE</b></font>

```
import numpy as np
from sklearn.metrics import mean_squared_error

y_pred = []
for i in range(no_base_models):
  sam_cols = list_selected_columns[i]
  sam_data = x[:,sam_cols]
  #print(sam_data.shape)
  model = list_all_models[i]
  y = model.predict(sam_data)
  #print(y.shape)
  y_pred.append(y)
y_pred = np.array(y_pred)
print((y_pred.shape))
y_pred= np.median(y_pred, axis= 0)
print((y_pred.shape))
mse = mean_squared_error(y,y_pred)
print(mse)
```

<font color='blue'><b>Step - 3 </b></font>

<font color='orange'><b>Flowchart for calculating OOB score</b></font>

![alt text](https://i.imgur.com/95S5Mtm.png)

Now calculate the $OOB Score =  \frac{1}{506}\sum_{i=1}^{506}(y^{i} - y^{i}_{pred})^{2}$.

*  <font color='blue'><b> Write code for calculating OOB score </b></font>

```
y_pred = []
for i in range(len(x)):
  sam=[]
  for j in range(no_base_models):
    if i not in list_selected_rows[j]:
      model = list_all_models[j]
      sam.append(model.predict(x[i,list_selected_columns[j]].reshape(1,-1)))
  y_pred.append(np.median(sam))

#print(len(y_pred))
oobscore = mean_squared_error(y,y_pred)
print(oobscore)

```

# <font color='blue'><b>Task 2</b></font>

```
Train_mse = []
OOb_score = []
no_iter = 35
for count2 in range(no_iter):
  list_input_data =[]
  list_output_data =[]
  list_selected_rows= []
  list_selected_columns=[]
  no_base_models = 30
  count = 0
  i=0
  while (i != (no_base_models+count)):
    a,b,c,d = generating_samples(x, y)
    d = sorted(d)
    i = i+1
    if ((d in list_selected_columns)):
      #print(count)
      count = count+1
      continue
    list_selected_columns.append(d)
    list_input_data.append(a)
    list_output_data.append(b)
    list_selected_rows.append(c)

  #print(len(list_input_data), len(list_selected_columns ))
  list_all_models = []
  i=0
  for i in range(no_base_models):
    input_data = list_input_data[i]
    target_data = list_output_data[i]
    model = DecisionTreeRegressor()
    model.fit(input_data, target_data)
    list_all_models.append(model)

  y_pred = []
  for i in range(no_base_models):
    sam_cols = list_selected_columns[i]
    sam_data = x[:,sam_cols]
    #print(sam_data.shape)
    model = list_all_models[i]
    y_pred.append(model.predict(sam_data))

  y_pred = np.array(y_pred)
  #print((y_pred.shape))
  y_pred= np.median(y_pred, axis= 0)
  #print((y_pred.shape))
  mse = mean_squared_error(y,y_pred)
  #print(mse)
  Train_mse.append(mse)

  y_pred = []
  for i in range(len(x)):
    sam=[]
    for j in range(no_base_models):
      if i not in list_selected_rows[j]:
        model = list_all_models[j]
        sam.append(model.predict(x[i,list_selected_columns[j]].reshape(1,-1)))
    y_pred.append(np.median(sam))

  #print(len(y_pred))
  oobscore = mean_squared_error(y,y_pred)
  print(mse,oobscore)
  OOb_score.append(oobscore)

print(len(Train_mse))
print(len(OOb_score))
```

```
print((Train_mse))
print((OOb_score))
```

```
x_mean = np.mean(Train_mse)
x_std = np.std(Train_mse)
#print(x_mean, x_std)
# for 95% confidence interval
upper_lim = x_mean-2*(x_std/np.sqrt(len(Train_mse)))
lower_lim = x_mean+2*(x_std/np.sqrt(len(Train_mse)))
#print(upper_lim,lower_lim)
print('95% confidence interval of mean square error is [{0},{1}]'.format(upper_lim,lower_lim))
```

```
x_mean = np.mean(OOb_score)
x_std = np.std(OOb_score)
#print(x_mean, x_std)
# for 95% confidence interval
upper_lim = x_mean-2*(x_std/np.sqrt(len(OOb_score)))
lower_lim = x_mean+2*(x_std/np.sqrt(len(OOb_score)))
#print(upper_lim,lower_lim)
print('95% confidence interval for Out of Bag(OOB) score is [{0},{1}]'.format(upper_lim,lower_lim))
```
